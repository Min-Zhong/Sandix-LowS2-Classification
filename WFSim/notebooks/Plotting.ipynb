#  Plotting

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Fix peak_gap_threshold and change g2 

```python
different_g2 = [5, 10, 20, 30, 40]
s1_rate_strax_different_g2 = [0.9003, 0.9925, 0.9965, 0.9975, 0.9975]
s1_rate_bdt_different_g2 = [0.9666, 0.9675, 0.994, 0.999, 1.00]
s1_rate_cut_different_g2 = [0.9727, 0.9845, 0.9970, 0.9885, 0.9995]
s2_rate_strax_different_g2 = [0.0917, 0.2319, 0.5278, 0.6161, 0.6502]
s2_rate_bdt_different_g2 = [0.763, 0.9345, 0.984, 0.9974, 0.9981]
s2_rate_cut_different_g2 = [0.6639, 0.8182, 0.9476, 0.9914, 0.9974 ]
```

```python
fig=plt.figure(figsize=(8,6))
fig.patch.set_color('white')
plt.plot(different_g2, s1_rate_strax_different_g2, color='r', marker='o', label='strax classification')
plt.plot(different_g2, s1_rate_bdt_different_g2, color='b', marker='*', label='BDT classification')
plt.plot(different_g2, s1_rate_cut_different_g2, color='g', marker='^', label='Simple cut on one feature')
plt.xlim(0,50)
plt.ylim(0.9,1.05)
plt.hlines(1, 0, 50, linestyle=':')
plt.legend(loc='lower right')
plt.xlabel('g2')
plt.ylabel('Rate')
plt.title('S1 Correct Classification Rate')
plt.show()
```

```python
fig=plt.figure(figsize=(8,6))
fig.patch.set_color('white')
plt.plot(different_g2, s2_rate_strax_different_g2, color='r', marker='o', label='strax classification')
plt.plot(different_g2, s2_rate_bdt_different_g2, color='b', marker='*', label='BDT classification')
plt.plot(different_g2, s2_rate_cut_different_g2, color='g', marker='^', label='Simple cut on one feature')
plt.xlim(0,50)
plt.ylim(0,1.1)
plt.hlines(1, 0, 50, linestyle=':')
plt.legend(loc='lower right')
plt.xlabel('g2')
plt.ylabel('Rate')
plt.title('S2 Correct Classification Rate')
plt.show()
```
